# Loading Fine-tuned Model



```python
if torch.cuda.is_available():
    model_id = "asmaa-ali/llama-2-7b-AsmaaChemProt"
    model = AutoModelForCausalLM.from_pretrained(model_id, torch_dtype=torch.float16, device_map="auto")
    model.cuda()
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    tokenizer.use_default_system_prompt = False
```

# Using Fine-tuned Model for Prediction

```python
from transformers import pipeline

print('---------------------------------------------------------------------------------')
print('-----Welcome to the ChemProtBot. You are now ready to start asking questions?----')
print('---------------------------------------------------------------------------------')

pipe = pipeline(task="text-generation", model=model,
                tokenizer=tokenizer, max_length=500,
                do_sample=True, temperature=0.9,
                top_p=0.6, truncation=True)

while True:
    query = input(f"\nPrompt: ")

    if query.lower() in {"exit", "quit", "q", "f"}:
        print('Exiting')
        break

    if query.strip() == '':
        continue

    result = pipe(f"<s>[INST] {query} [/INST]")
    print(f"\nAnswer: " + result[0]['generated_text'].split('[/INST]  ')[1])
```
